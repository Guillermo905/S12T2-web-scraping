Realitza web scraping d'una pàgina de la borsa de Madrid (https://www.bolsamadrid.es) utilitzant BeautifulSoup i Selenium.

```python
import requests

URL = 'https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000'

page = requests.get(URL)

print(page.text)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
results = soup.find(id="BuscadorCab")
```

```python
print(results)
```

```python
print(results.prettify())
```

```python
len(results)
```

```python
elements = soup.find_all('div', class_ = 'noimpr')
```

```python
len(elements)
```

```python
elements2 = soup.find_all('div', id = 'Base')
```

```python
print(elements2)
```

```python
#Es mostren els diferents elements separats per paràgrafs
for element in elements:
    print(element, end="\n"*2)
```

```python
#S'intenta filtrar una mica més el codi
for element in elements2:
    menu = element.find('td',class_="BaseMenu noimpr")
    
```
